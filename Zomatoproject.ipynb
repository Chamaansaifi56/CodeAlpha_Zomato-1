QUESTION 2.

How many votes has each type of restaurant recevied from customers?

```python
df.head()
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result,c= "green",marker = "o")
plt.xlabel("Type of restaurant",c ="red", size = 20)
plt.ylabel("votes",c = "red", size = 20)
```

conclusion -  dinning resturants has recieved maximum votes

QUESTION 6.

Which type of resturant received more offline orders, so that Zomoto can rating coutomers with some good offers?

```python
df.head()
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order',aggfunc = 'size',fill_value= 0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed in (Type)")
plt.show()

```

Conclusion - Dining restaurants primarily accept orders, where as cafes primarily recive online orders. This suggests that clients prefers orders in person at restaurants, but prefer online ordering at cafes.
